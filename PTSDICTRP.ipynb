```python
import xml.etree.ElementTree as ET
import xlsxwriter as xlsx
import fnmatch
import os
import time
import pandas as pd
from collections import Counter 
import openpyxl
import csv

start = time.time()
file = r'C:\Users\Billy.Peacock\Desktop\PythonWebScraping\ICTRP-Results.xml'

workbook = xlsx.Workbook('PTSDICTRP1.xlsx')

worksheetnames = workbook.add_worksheet("Names")
worksheetnonames = workbook.add_worksheet("NoNames")
worksheetcompare = workbook.add_worksheet("Compare")

worksheetnames.write(0, 0, "Messy_Name")
worksheetnames.write(0, 1, "Official_Name")
worksheetnames.write(0, 2, "First_Name")
worksheetnames.write(0, 3, "Middle_Name")
worksheetnames.write(0, 4, "Last_name")
worksheetnames.write(0, 5, "Official_Role")
worksheetnames.write(0, 6, "Official_Affil")
worksheetnames.write(0, 7, "ID")
worksheetnames.write(0, 8, "Official_Title")
worksheetnames.write(0, 9, "Condition")
worksheetnames.write(0, 10, "Intervention_Name")
worksheetnames.write(0, 11, "Agency")
worksheetnames.write(0, 12, "Phase")
worksheetnames.write(0, 13, "Start_Date")
worksheetnames.write(0, 14, "Completion_Date")
worksheetnames.write(0, 15, "Last_Updated")
worksheetnames.write(0, 16, "Location")
worksheetnames.write(0, 17, "URL")
worksheetnames.write(0, 18, "Registry")


worksheetnonames.write(0, 0, "ID")
worksheetnonames.write(0, 1, "Official_Title")
worksheetnonames.write(0, 2, "Condition")
worksheetnonames.write(0, 3, "Intervention_Name")
worksheetnonames.write(0, 4, "Agency")
worksheetnonames.write(0, 5, "Phase")
worksheetnonames.write(0, 6, "Start_Date")
worksheetnonames.write(0, 7, "Completion_Date")
worksheetnonames.write(0, 8, "Last_Updated")
worksheetnonames.write(0, 9, "Location")
worksheetnonames.write(0, 10, "URL")
worksheetnonames.write(0, 11, "File_Ref")

worksheetcompare.write(0,0, "NZ_Trials")
worksheetcompare.write(0,1, "Old_Missing")
worksheetcompare.write(0,2, "Sorted_New")
worksheetcompare.write(0,3, "Sorted_Old")


row = 1
col = 0
row1 = 1
col1 = 0

all_params = []

tree = ET.parse(file)
root = tree.getroot()

idcheck = ['NCT', 'ACTRN', 'DRKS']

for trial in root.findall('Trial'):
    ID = trial.find('TrialID').text
    
    if any(x in ID for x in idcheck):
        pass
    else:
        messyname= ''
        all_params.append(messyname)
        
        offname = ''
        all_params.append(offname)
        
        firstname = trial.find('Contact_Firstname')
        if firstname is not None:
            firstname = firstname.text
            all_params.append(firstname)
        else:
            all_params.append(' ')
        
        mname = ''
        all_params.append(mname)
        
        lastname = trial.find('Contact_Lastname')
        if lastname is not None:
            lastname = lastname.text
            all_params.append(lastname)
        else:
            all_params.append(' ')
        
        if 'ChiCTR' in ID:
            all_params.append('Study Leader')
        elif 'KCT' in ID:
            all_params.append('Principle Investigator')
        elif 'IRCT' in ID:
            all_params.append('General Enquiries')
        elif 'NL' in ID:
            all_params.append('Contact')
        elif 'NTR' in ID:
            all_params.append('Contact')
        else:
            role = ' '
            all_params.append(role)
        
        affil = trial.find('Contact_Affiliation')
        if affil is not None:
            affil = affil.text
            all_params.append(affil)
        else:
            all_params.append(' ')
        
        all_params.append(ID)
        
        stitle = trial.find('Scientific_title')
        locked = stitle.attrib
        
        if locked.get('{http://www.w3.org/XML/1998/namespace}space') == 'preserve':
            ptitle = trial.find('Public_title')
            all_params.append(ptitle.text)
        else:
            all_params.append(stitle.text)
        
        condition = trial.find('Condition')
        if condition is not None:
            condition = condition.text
            all_params.append(condition)
        else:
            all_params.append(' ')   
        
        intervention = ' '
        all_params.append(intervention)   
        
        sponsor = trial.find('Primary_sponsor')
        if sponsor is not None:
            sponsor = sponsor.text
            all_params.append(sponsor)
        else:
            all_params.append(' ')
        
        phase = trial.find('Phase')
        if phase is not None:
            phase = phase.text
            all_params.append(phase)
        else:
            all_params.append(' ')
        
        startdate = trial.find('Date_registration')
        if startdate is not None:
            startdate = startdate.text
            all_params.append(startdate)
        else:
            all_params.append(' ')
            
        enddate = ' '
        all_params.append(enddate)
        
        lastupdate = trial.find('Last_Refreshed_on')
        if lastupdate is not None:
            lastupdate = lastupdate.text
            all_params.append(lastupdate)
        else:
            all_params.append(' ')
        
        location = ' '
        all_params.append(location)
        
        url = trial.find('results_url_link')
        if url is not None:
            url = url.text
            all_params.append(url)
        else:
            all_params.append(' ')
        
        registry = trial.find('Source_Register')
        if registry is not None:
            registry = registry.text
            all_params.append(registry)
        else:
            all_params.append(' ')  
            
        #print(all_params)
        
        for i in all_params:
            worksheetnames.write(row, col, i)
            col += 1
        row+=1
        col = 0
                
        all_params = []
                
worksheetnames.set_column(0,20,20)
worksheetnonames.set_column(0,20,20)
worksheetcompare.set_column(0,20,20)

workbook.close()

end = time.time()
print("Finished in", end-start, "Seconds")
```
